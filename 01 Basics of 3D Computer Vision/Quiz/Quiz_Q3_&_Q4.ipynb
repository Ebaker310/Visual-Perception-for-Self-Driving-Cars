```
import numpy as np
import math

cos = math.cos(math.pi)
sin = math.sin(math.pi)

# projection from world -> camera
# oImage = [R|t]oWorld

oWorld = np.array([[-0.5, 1.5, 9, 1]])
Rt = np.array([[1, 0, 0, 1], [0, cos, -sin, 2], [0, sin, cos, 10]])

oCamera = np.dot(Rt, oWorld.T)
print(oCamera)
```

```
# oImage = K * oCamera
K = np.array([[640, 0 , 640], [0, 480, 480], [0, 0, 1]])
oImage = np.dot(K, oCamera)
print("oImage = %d,%d"%(oImage[0],oImage[1]))


```
